```python
import google.generativeai as genai
from IPython.display import Markdown

# 1. SETUP
GOOGLE_API_KEY = "PASTE_YOUR_API_KEY_HERE"
genai.configure(api_key=GOOGLE_API_KEY)

# 2. FIND THE RIGHT MODEL AUTOMATICALLY
valid_model = None
for m in genai.list_models():
  if 'generateContent' in m.supported_generation_methods:
    if 'gemini' in m.name:
      valid_model = m.name
      print(f"Found working model: {valid_model}")
      break

if not valid_model:
    print("Could not auto-detect. Trying default 'gemini-1.5-flash'...")
    valid_model = 'models/gemini-1.5-flash'

model = genai.GenerativeModel(valid_model)

# 3. FUNCTION TO GENERATE ROADMAP
def generate_career_roadmap(current_skills, target_role):
  print(f"ğŸš€ Generating roadmap using {valid_model}...")

  prompt = f"""
  Act as a Senior Career Mentor.
  My Goal: Become a {target_role}.
  My Current Skills: {current_skills}.

  Create a 4-week structured learning roadmap to bridge the gap.
  Format the output nicely with emojis and bullet points.
  For each week, suggest 1 specific project I should build.
  """

  try:
    response = model.generate_content(prompt)
    return response.text
  except Exception as e:
    return f"Error occurred: {str(e)}"

# --- USER INTERFACE ---
print("\n--- ğŸ“ CareerPath AI Prototype ---")
user_skills = input("Enter your current skills (e.g., HTML, CSS): ")
user_target = input("Enter your dream job (e.g., MERN stack Developer): ")

roadmap = generate_career_roadmap(user_skills, user_target)

print("\n--- âœ… YOUR PERSONALIZED ROADMAP ---")
Markdown(roadmap)
```
